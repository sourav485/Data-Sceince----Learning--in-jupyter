```python
import matplotlib.pyplot as plt 

x = [100, 1200, 2200, 500, 3400, 8050, 40000]
y = ['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun']

plt.title('sales of the month')

plt.xlabel('sales per week')
plt.ylabel('weeks per week')

plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt 

months = [1, 2, 3, 4]
sales = [1000, 2000, 1200, 3000]

plt.plot(months, sales, color = 'blue', linestyle = '--', linewidth = 2, marker = 'o', label = 'sales report 2025')

plt.title('monthly sales data report')
plt.xlabel('months')
plt.ylabel('sales of 2025')
plt.legend(loc = 'upper left', fontsize = 10)
plt.grid(color = 'grey', linestyle = ':', linewidth = 1)
plt.xlim(1, 4)
plt.ylim(1000, 3000)
plt.xticks([1, 2, 3, 4], ['M1', 'M2', 'M3', 'M4']  )
plt.show()
```

```python
import matplotlib.pyplot as plt

days = ['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun']
temp = [10, 20, 12, 13, 34, 45, 19]

plt.plot(days, temp, marker = 'o', color = 'blue', linestyle = '--', label = 'tempereture in week' )
plt.xlabel('days')
plt.ylabel('tempereture')
plt.title('tempreture in a week')
plt.legend(loc = 'upper left', fontsize = 10)
plt.grid(color = 'grey')

plt.show()
```

```python
import matplotlib.pyplot as plt

Wipro = [100, 110, 100, 130, 120, 155, 145]
ITC = [110, 130, 140, 150, 120, 130, 135]
Day = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

plt.plot(Day, ITC, Wipro, color = 'blue', marker = 'o', linestyle = '--', linewidth = 2, label = 'share price in a week')
plt.title('Share price chart')
plt.xlabel('day')
plt.ylabel('price')
plt.grid(color = 'green')
plt.legend(loc = 'upper left', fontsize = 10)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)
# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
import matplotlib.pyplot as plt

subjects = ['Math', 'Science', 'English', 'History']
john = [85, 90, 80, 75]
emma = [78, 88, 92, 70]
plt.plot(subjects, john, label='John', marker='o')
plt.plot(subjects, emma, label='Emma', marker='s')
plt.title('Student Performance Comparison')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

product = ['A', 'B', 'C', 'D']
sales = [1000, 800, 1200, 1600]

plt.bar(product, sales, label = 'sales in 2025', color = 'orange')
plt.xlabel('product')
plt.ylabel('sales')
plt.title('sales in diff prodects')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

region = ['north', 'south', 'east', 'west']
sales = [1000, 1200, 2300, 1300]

plt.pie(sales, labels = region, autopct = '%1.1f%%', colors = ['skyblue', 'gold', 'green', 'orange'])
plt.title('sales distribution among different regions')
plt.show()
```
